```python
import pandas as pd
import itertools as it
import numpy as np
```

```python
def get_footer(file_):
    with open(file_) as f:
        g = it.dropwhile(lambda x: x != 'variableStep chrom=contig_2 span=1\n', f)
        footer_len = len([i for i, _ in enumerate(g)])
    return footer_len
```
