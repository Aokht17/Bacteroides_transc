```python
import pandas as pd
import re
```

```python
data = pd.read_csv('b_uni_type.gff', comment='#', sep='\t', skipfooter=78531, names=['ref', 'prodigal', 'CDS', 'start', 'end', 'row', 'strand', 'row2', 'annot'], engine='python')
data = data[['annot']]
data
```

```python
transc_dict = {}
for _,row in data.iterrows():
    text=row.annot
    ID = re.compile(r'ID=([A-Za-z0-9_-]+)') 
    name = re.compile(r'Name=([A-Za-z0-9_-]+)') 
    findid = ID.findall(text.strip())[0]
    try:
        findname = name.findall(text.strip())[0]
    except:
        findname = findid if 'ncRNA' not in findid else findid.split('_')[2]
        
    transc_dict.update({findid:findname})
```

```python
transc_dict
```

```python
id_name = pd.DataFrame.from_dict(transc_dict, orient='index', columns=['gene'])
id_name = id_name.reset_index()
id_name
```

```python
id_name.to_csv('gene_names.tsv', sep='\t', index=False)
```

```python
data1 = pd.read_csv('b_uni_type.gff', comment='#', sep='\t', skipfooter=78531, names=['ref', 'prodigal', 'CDS', 'start', 'end', 'row', 'strand', 'row2', 'annot'], engine='python')
data1 = data1[['start', 'end', 'annot']]
data1
```

```python
data1['len'] = data1.end-data1.start+1
data1
```

```python
data1 = data1[['annot', 'len']]
```

```python
len_dict = {}
for _,row in data1.iterrows():
    text=row.annot
    ID = re.compile(r'ID=([A-Za-z0-9_-]+)') 
    name = re.compile(r'Name=([A-Za-z0-9_-]+)') 
    findid = ID.findall(text.strip())[0]
    try:
        findname = name.findall(text.strip())[0]
    except:
        findname = 'undefined' if 'ncRNA' not in findid else findid.split('_')[2]
        
    len_dict.update({findid:row.len})
```

```python
len_dict
```

```python
id_name['len']= id_name['index'].map(len_dict)
id_name
```

```python
id_name[['gene', 'len']].to_csv('gene_len.tsv', sep='\t', index=False)
```

```python
data = pd.read_csv('../down/b_theta_genomic.gff', comment='#', sep='\t', names=['ref', 'prodigal', 'CDS', 'start', 'end', 'row', 'strand', 'row2', 'annot'], engine='python')
data = data[['CDS','annot']]
data = data[data.CDS == 'CDS']
data=data[['annot']]
data
```

```python
transc_dict = {}
for _,row in data.iterrows():
    text=row.annot
    ID = re.compile(r'ID=([A-Za-z0-9_.-]+)') 
    name = re.compile(r'gene=([A-Za-z0-9_-]+)') 
    BT = re.compile(r'Parent=([A-Za-z0-9_-]+)') 
    findid = ID.findall(text.strip())[0]
    findbt = BT.findall(text.strip())[0]
    try:
        findname = name.findall(text.strip())[0]
    except:
        findname = findid
    transc_dict.update({findbt:findname})
```

```python
transc_dict
```

```python
id_name = pd.DataFrame.from_dict(transc_dict, orient='index', columns=['gene'])
id_name = id_name.reset_index()
id_name
```

```python
id_name.gene = id_name.gene.str.strip('cds-')
id_name
```

```python
id_name.to_csv('gene_names_Lucia_2.tsv', sep='\t', index=False)
```
