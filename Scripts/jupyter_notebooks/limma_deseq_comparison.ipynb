```python
tr = pd.read_csv('/home/nastya/Bacteroides_uniformis_annotations/DEG_data/stat_vs_exp_DEG_results.csv')
tr
```

```python
tr = tr[['Unnamed: 0', 'log2FoldChange', 'padj']]
tr = tr.rename(columns={'Unnamed: 0': 'Gene'})
tr
```

```python
signif_tr = tr[(abs(tr.log2FoldChange) > 0.5) & (tr. padj < 0.05)]
signif_tr
```

```python
limma = pd.read_csv('/home/nastya/Bacteroides_uniformis_annotations/proteomics/S3421_LM10C_analysis_result.csv')
limma = limma[limma.comparison == 'exp_phase-sta_phase'][['Gene','logFC', 'adj.P.Val' ]]
```

```python
signif_limma = limma[(abs(limma.logFC) > 0.5) & (limma['adj.P.Val'] < 0.05)]
signif_limma
```

```python
a = pd.merge(signif_limma, signif_tr)
a = a.rename(columns={'logFC': 'logFC_prot', 'adj.P.Val': 'padj_prot', 'log2FoldChange': 'logFC_tr', 'padj': 'padj_tr'})
a.logFC_prot = a.logFC_prot*(-1)
a
```

```python
a.to_csv('stat_VS_exp_prot_and_tr.csv')
```
