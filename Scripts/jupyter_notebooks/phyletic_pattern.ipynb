```python
path = '/home/nastya/Downloads/bacteroides_50_gff/ncbi-genomes-2023-02-14/'
genes = ['dinB', 'recA', 'ruvA', 'polA', 'recN', 'nusA', 'uvrA', 'uvrB', 'dnaK']
files = [path+i for i in os.listdir(path)]
for gff in files:
    with open(gff, "r") as file:
        for lines in file:
            line = file.readline()
            for gene in genes:
                if gene in line:
                    with open('phyletic_pattern.txt', 'a') as out:
                        a = gff.split('/')[6]
                        name = a.split('_')[0]+'_'+a.split('_')[1]
                        out.write(f"{name}\t{gene}\n")
```
