```python
import re
import numpy as np
```

```python
data = pd.read_csv('b_theta.gff3', comment='#', sep='\t', names=['ref', 'prodigal', 'CDS', 'start', 'end', 'row', 'strand', 'row2', 'annot'], engine='python')
data
```

```python
data.CDS.unique()
```

```python
data = data[data.CDS == 'CDS']
data = data[['annot']]
data
```

```python
transc_dict = {}
for _,row in data.iterrows():
    text=row.annot
    ID = re.compile(r'ID=cds-([.A-Za-z0-9_-]+)') 
    name = re.compile(r'Parent=gene-([,:A-Za-z0-9_-]+)') 
    findid = ID.findall(text.strip())[0]
    try:
        findname = name.findall(text.strip())[0]
    except:
        findname = np.nan
        
    transc_dict.update({findid:findname})
    
transc_dict
```

```python
wpbt = pd.DataFrame.from_dict(transc_dict, orient='index', columns=['btrs'])
wpbt = wpbt.reset_index()
wpbt
```

```python
wpbt = wpbt.rename(columns={'index': 'wp'})
wpbt
```

```python
transc_dict = {}
for _,row in data.iterrows():
    text=row.annot
    ID = re.compile(r'ID=cds-([.A-Za-z0-9_-]+)') 
    name = re.compile(r'product=([ ,:A-Za-z0-9_-]+)') 
    findid = ID.findall(text.strip())[0]
    try:
        findname = name.findall(text.strip())[0]
    except:
        findname = np.nan
        
    transc_dict.update({findid:findname})
    
transc_dict
```

```python
wpprod = pd.DataFrame.from_dict(transc_dict, orient='index', columns=['product'])
wpprod = wpprod.reset_index()
wpprod = wpprod.rename(columns={'index': 'wp'})
wpprod
```
