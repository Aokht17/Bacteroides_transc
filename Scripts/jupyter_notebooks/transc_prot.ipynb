```python
data = pd.read_csv('./metatranscriptomics/b_uni_type.gff', comment='#', sep='\t', skipfooter=78531, names=['ref', 'prodigal', 'CDS', 'start', 'end', 'row', 'strand', 'row2', 'annot'], engine='python')
data = data[['annot']]

transc_dict = {}
for _,row in data.iterrows():
    text=row.annot
    ID = re.compile(r'ID=([A-Za-z0-9_-]+)') 
    name = re.compile(r'eggNOG=411479.([A-Za-z0-9_-]+)') 
    findid = ID.findall(text.strip())[0]
    try:
        findname = name.findall(text.strip())[0]
    except:
        findname = findid
        
    transc_dict.update({findid:findname})
    
transc_dict
```

```python
id_name = pd.DataFrame.from_dict(transc_dict, orient='index', columns=['gene'])
id_name = id_name.reset_index()
id_name
```

```python
egg = pd.read_csv('metatranscriptomics/eggnog.csv', comment='#')[['seed_ortholog', 'Preferred_name']]
egg
```

```python
egg[['1', 'bacuni']]=egg['seed_ortholog'].str.split(".", expand = True)
egg = egg[['bacuni', 'Preferred_name']]
egg
```

```python
for _,row in s.iterrows():
    if type(row['Gene']) == float:
        print(row['Gene'])
```

```python
final = pd.DataFrame(columns = ['index', 'bacuni', 'Entry.Name', 'Gene'])
for _,row in s.iterrows():
    if type(row['Gene']) == float:
        try:
            a = egg[egg['bacuni']==row['gene']].Preferred_name.item()
            symbol = a if type(a) != float else row['gene']
        except:
            symbol = row['gene']
        final = final.append({'index' : row['index'], 'bacuni' : row['gene'], 'Entry.Name' : row['Entry.Name'], 'Gene' : symbol},ignore_index = True)
    else:
        final = final.append({'index' : row['index'], 'bacuni' : row['gene'], 'Entry.Name' : row['Entry.Name'], 'Gene' : row['Gene']},ignore_index = True)
```
