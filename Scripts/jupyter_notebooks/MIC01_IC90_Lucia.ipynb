```r
library(dplyr)
library(drc)  # Dose Response Curves
```

```r
dat <- read.csv(file = 'df5.csv')
```

```r
hist(dat$relative_growt)
```

```r
dat
```

```r
df = NULL
for (strain in unique(dat$strain)) {
    print(strain)
    for (drug in unique(dat$drug)) {
        print(drug)
        dat3 = dat[(dat$strain==strain & dat$drug==drug),]
        fit = drm(relative_growth ~ conc, data = dat3, fct = LL2.4())
        mic90 = ED(fit, 90, interval = 'fls')
        plot(fit, type = 'all')
        abline(v = mic90, lty = c(1, 2, 2))  #mic90 with CI95%
        mic90 = data.frame(mic90)
        Estimate_IC90 = mic90$Estimate
        Lower_IC90 = mic90$Lower
        Upper_IC90 = mic90$Upper
        df = rbind(df, data.frame(strain,drug,Estimate_IC90, Lower_IC90, Upper_IC90))
    }
}
```

```r
df$confidance_interval_range = df$Upper_IC90/df$Lower_IC90
```

```r
hist(df$confidance_interval_range)
```

```r
write.csv(df, file = "df_IC90.csv")
```
