```python
uniprot=pd.read_csv('uniprot_annot.tsv', sep='\t')[['Protein names', 'Gene Names', 'Gene Ontology (biological process)', 'Gene Ontology (GO)']]
uniprot
```

```python
uniprot['bacuni'] = uniprot['Gene Names'].apply(lambda x: re.search(r'BACUNI_[0-9]*', str(x)).group() if 'BACUNI' in str(x) else np.nan)
```

```python
uniprot = uniprot[['bacuni', 'Gene Ontology (biological process)']]
uniprot
```

```python
uniprot[['go1', 'go2', 'go3', 'go4', 'go5', 'go6', 'go7']]=uniprot['Gene Ontology (biological process)'].str.split(";", expand = True)
uniprot
```

```python
uniprot['go_1'] = uniprot['go1'].apply(lambda x: re.search(r"GO:[0-9]*", str(x)).group() if 'GO' in str(x) else np.nan)

uniprot['go_2']= uniprot['go2'].apply(lambda x: re.search(r"GO:[0-9]*", str(x)).group() if 'GO' in str(x) else np.nan)

uniprot['go_3']= uniprot['go3'].apply(lambda x: re.search(r"GO:[0-9]*", str(x)).group() if 'GO' in str(x) else np.nan)
uniprot['go_4']=uniprot['go4'].apply(lambda x: re.search(r"GO:[0-9]*", str(x)).group() if 'GO' in str(x) else np.nan)
uniprot['go_5']=uniprot['go5'].apply(lambda x: re.search(r"GO:[0-9]*", str(x)).group() if 'GO' in str(x) else np.nan)
uniprot['go_6']=uniprot['go6'].apply(lambda x: re.search(r"GO:[0-9]*", str(x)).group() if 'GO' in str(x) else np.nan)
uniprot['go_7']=uniprot['go7'].apply(lambda x: re.search(r"GO:[0-9]*", str(x)).group() if 'GO' in str(x) else np.nan)

uniprot['go']=uniprot['go_1'].fillna('')+','+uniprot['go_2'].fillna('')+','+uniprot['go_3'].fillna('')+','+uniprot['go_4'].fillna('')+','+uniprot['go_5'].fillna('')+','+uniprot['go_6'].fillna('')+','+uniprot['go_7'].fillna('')
uniprot
```
