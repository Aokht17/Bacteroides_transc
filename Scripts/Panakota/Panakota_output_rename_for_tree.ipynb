```python
data = data[['gembase_name', 'orig_name']]
data
```

```python
data['orig_name'] = data['orig_name'].str.strip('Bacteroides/tmp_files/')
data
```

```python
data[['GCF', 'num', '1', '2', '3']] = data['orig_name'].str.split('_', expand=True)
data
```

```python
data['num'] = 'GCF_' + data['num']
data
```

```python
data = data.drop(['orig_name', '1', '2', '3', 'GCF'], axis=1)
```

```python
data
```

```python
import os
import re
import requests
from bs4 import BeautifulSoup
```

```python
assembly_list = data['num'].to_list()
```

```python
data['annot'] = np.NaN
```

```python
data
```

```python
annot_list = []
for assembly in assembly_list:
    try:
        url = "https://www.ncbi.nlm.nih.gov/assembly/{}"
        r = requests.get(url.format(assembly))  # Access url page of input platform
        parse_data = r.text  # Get whole page in HTML format
        soup = str(BeautifulSoup(parse_data, "html.parser"))  # Pull data from HTML format
        req = re.search(
            r"(?<=Species name</th></tr><tr><td>)[a-zA-Z0-9 ]*(?=</td>)", soup, re.DOTALL | re.MULTILINE
        )  # Get data from Title field
        result = ""  # Data extracted from req and/or req1 fields will be put into result
        if req:
            result = str(req.group())
        entry = result.split(' ')[0]+'_'+result.split(' ')[1]+'_'+assembly
        annot_list.append(entry)
    except:
        annot_list.append(np.NaN)
        print(assembly)
```

```python
data['annot'] = annot_list
data
```
